```python
import pandas as pd
```

```python
path = '../Sesion2/data_retail'

path_customers = f'{path}/customers'
path_orders = f'{path}/orders'
path_order_items = f'{path}/order_items'
path_products = f'{path}/products'
path_categories = f'{path}/categories'
path_departments = f'{path}/departments'
```

```python
df_customers = pd.read_csv(path_customers, sep='|', header = None)
df_orders = pd.read_csv(path_orders, sep='|', header = None)
df_order_items = pd.read_csv(path_order_items, sep='|', header = None)
df_products = pd.read_csv(path_products, sep='|', header = None)
df_categories = pd.read_csv(path_categories, sep='|', header = None, names=['category_id', 'category_department_id', 'category_name'])
df_departments = pd.read_csv(path_departments, sep='|', header = None, names=['department_id', 'department_name'])
```

```python
def validate_duplicate(df, column_name ):
    if df[column_name].duplicated().any():
        print(f"Hay duplicados en el dataframe en la columna {column_name}")
        exit(1)
    else:
        print("No existe duplicados")
```

```python
def validate_department(df):
    if df['department_name'].duplicated().any():
        print("Hay departamentos duplicados en el dataframe")
        exit(1)
    else:
        print("No existe duplicados")
```

```python
validate_department(df_departments)
```

```python
validate_duplicate(df_departments, 'department_name')
```

```python
validate_duplicate(df_categories, 'category_department_id')
```

```python
def validate_ids(df,valid_ids, column_name):
    if not df[column_name].isin(valid_ids).all():
         print("Hay ids que no se encuentran en el dataframe")
    else:
         print("OK")
```

```python
valid_ids = set(df_departments['department_id'])
valid_ids
```

```python
if not df_categories['category_department_id'].isin(valid_ids).all():
    print("Hay ids que no se encuentran en el dataframe")
```

```python
validate_ids(df_categories, valid_ids, 'category_department_id')
```
